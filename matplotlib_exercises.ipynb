```python
# Exercise 3
# Combine the figures you created in the last step into one large figure with 4 subplots.



x1 = np.array(range(0, 100))
y1 = x1**2 - x1 + 2

plt.subplot(2,2,1)
plt.plot(x1,y1, color = 'green')
plt.title('y = x^2 âˆ’ x + 2')




x2 = np.arange(0, 25)
y2 = np.sqrt(x2)

plt.subplot(2,2,2)
plt.plot(x2,y2, color = 'red')
plt.title('y = âˆš x')


x3 = np.arange(-25, 25)
y3 = x3**3
   
plt.subplot(2,2,3)
plt.plot(x3,y3, color = 'yellow')
plt.title('x3**3')



x4 = np.arange(0, 50)
y4 = 2 ** x4
   
plt.subplot(2,2,4)
plt.plot(x4,y4, color = 'purple')
plt.title('2 ** x4')
```
